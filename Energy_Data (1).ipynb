```python
#import the Energy_data_set(inspect)
import pandas as pd
df=pd.read_excel(r'C:\Users\Humphery\Desktop\Energy_data.xlsx')
df
```

```python
#check for missing_values in  the Energy data set
df.isnull().sum()
```

```python
# drop date and lights in the Energy data set
df=df.drop(columns=['date','lights'])
df
```

```python
#select Temperature outside the Building (northside) (T6), against Temperature in the living room area (T2) to show the simple linear regression.
simple_linear_reg_df=df[['T2', 'T6']].sample(n=15,random_state=2) 

import seaborn as sns    # to show the linear_Regression 
sns.regplot(x='T2',y='T6' ,data=simple_linear_reg_df) 
```

```python
# normalize the Energy_data(to form dummy_variables)
from sklearn .preprocessing import MinMaxScaler
scaler=MinMaxScaler()
normalized_data=pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
normalized_data
```

```python
#Features data(independent variables) .
features_names=normalized_data.drop(columns=['Appliances'])
features_names
```

```python
#inspect the target_variable(dependent _variable)
target_var=normalized_data[['Appliances']]
target_var
```

```python
#inspect the Energy_data using linear regression model by training and testing the linear_regression model.
from sklearn .linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features_names,target_var,test_size=0.3,random_state=42)
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
prediction=linear_model.predict(x_test)
#Exmine linear model coefficents
linear_model.coef_
```

```python
# mean absolute error of the Energy_data(mae)
from sklearn .metrics import mean_absolute_error
mae=mean_absolute_error(y_test,prediction)
mae
round(mae,2)# round to three decimal place
```

```python
#Residual Sum of square error of the Energy_data set(rss)
import numpy as np
rss=np.sum(np.square(y_test-prediction))
round(rss,2) #round to three decimal place
```

```python
#Root mean square_Error of the Energy_data set(rmse)
from sklearn.metrics import mean_squared_error
rms=np.sqrt(mean_squared_error(y_test,prediction))
round(rms,3)#round to 3 dp
```

```python
# the coefficient_of determination of the Energy data set(r2_score).( about 15% of the data_set is describe, 85% are affected by some external factors)
from sklearn .metrics import r2_score
r2_score=r2_score(y_test,prediction)
r2_score
round(r2_score,2)# round to 3 dp
```

```python
#inspecting the Energy _data set using Least  Absolute  Shrinkage and selection Operator(L1)
from sklearn .linear_model import Lasso
lasso_reg=Lasso(alpha=0.001)
lasso_reg.fit(x_train,y_train)
lasso_reg.coef_
```

```python
# inspecting the energy data set using Ridge Regression(L2)
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=0.5)
ridge_reg.fit(x_train,y_train)
ridge_reg.coef_
```

```python
# to detemine r2score of T6 against T2
#inspect the independent variable
temp_in_room=normalized_data[['T2']]
temp_in_room
```

```python
#inspect the dependent variable
temp_out_side=normalized_data[['T6']]
temp_out_side
```

```python
#the coefficent of detemination of Temperature outside the building(north side)(in celisus)(T6) and Temperature in the living_room area(T2).
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn .model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(temp_in_room, temp_out_side ,test_size=0.3,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
prediction2=model.predict(x_test)
prediction2
# deterime the coefficent of detemination of T6 against T2
r2_score=r2_score(y_test,prediction2)
r2_score
round(r2_score,2)
```

```python
#the Root mean square error using Least  Absolute  Shrinkage and selection Operator(L1) Regression
from sklearn .linear_model import Lasso
lasso_reg=Lasso(alpha=0.001)
lasso_reg.fit(x_train,y_train)
prediction3=lasso_reg.predict(x_test)
prediction3
```

```python
#root_mean square error using Lasso
from sklearn.metrics import mean_squared_error
import numpy as np
rmse=np.sqrt(mean_squared_error(y_test,prediction3))
round(rmse,3)
```
